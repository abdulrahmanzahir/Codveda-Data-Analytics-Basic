### Import needed Python libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style for plots
sns.set(style="whitegrid")

print("Libraries imported successfully")
```

## Step 2 â€” Load Cleaned Dataset

```python
# Load the cleaned dataset from Task 1
df = pd.read_csv("../Task 1 - Data Cleaning And Preprocessing/Cleaned-Dataset/iris_cleaned.csv")

# Show first 5 rows
df.head()
```

## Step 3 â€” Visualizations
### Step 3.1 â€” Species Count
A bar plot showing the number of samples per species in the cleaned dataset.

```python
# Create output directory for plots
outdir = "../Task 3 - Basic Data Visualization/Plots"
os.makedirs(outdir, exist_ok=True)

# Plot species count
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=df, palette="Set2")
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")

# Save and show
plt.savefig(f"{outdir}/species_count.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Step 3.2 â€” Feature Distributions
Histograms for each numerical feature to visualize their overall distribution.

```python
# Plot histograms for each feature
df.hist(figsize=(10,8), bins=15, edgecolor="black", color="#69b3a2")
plt.suptitle("Feature Distributions", fontsize=16)

# Save and show
plt.savefig(f"{outdir}/feature_distributions.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Step 3.3 â€” Pairplot by Species
A pairplot to visualize the relationships between all numerical features, color-coded by species.

```python
# Pairplot of all features grouped by species
sns.pairplot(df, hue="species", diag_kind="hist", palette="Set2")

# Save and show
plt.savefig(f"{outdir}/pairplot_species.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Step 3.4 â€” Boxplots by Species
Boxplots to compare the distribution of each feature across different species.

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

plt.figure(figsize=(12,8))
for i, feature in enumerate(features, 1):
    plt.subplot(2,2,i)
    sns.boxplot(x="species", y=feature, data=df, palette="Set2")
    plt.title(f"{feature} by Species")

plt.tight_layout()

# Save and show
plt.savefig(f"{outdir}/boxplots_species.png", dpi=300, bbox_inches="tight")
plt.show()
```

## ðŸ“ˆ Average Feature Values per Species  

This line chart shows the **mean values of each numeric feature** (`sepal_length`, `sepal_width`, `petal_length`, `petal_width`) for every Iris species.  
It provides a clear comparison of how the species differ in their average measurements across all features.  

```python
# Calculate mean values of features grouped by species
mean_values = df.groupby("species").mean()

# Plot line chart
plt.figure(figsize=(8,6))
for species in mean_values.index:
    plt.plot(mean_values.columns, mean_values.loc[species], marker="o", label=species)

plt.title("Average Feature Values per Species")
plt.xlabel("Features")
plt.ylabel("Mean Value")
plt.legend(title="Species")
plt.grid(True)

# Save and show
plt.savefig(f"{outdir}/lineplot_mean_features_per_species.png", dpi=300, bbox_inches="tight")
plt.show()
```
